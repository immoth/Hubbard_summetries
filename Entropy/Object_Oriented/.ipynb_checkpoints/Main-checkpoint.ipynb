```python
from Define_Optimizer import SPSA
from Define_Quantum_Device import Quantum_Device
import numpy as np
```

```python
qd = Quantum_Device()

def H_paulis(m,k):
    h = {'ZI':m,'IZ':m,'XX':k}
    return h

hpauli = H_paulis(1,0.3)
S0 = (0.0)*2*np.log(2)
phi0 = [[0.3,0.2],[0.5,0.1],[0.3,0.7],[1.2,0.9]]

T = 2.0
kmax = 300
El,kl,Sf,phif = SPSA(T, S0, phi0, kmax, hpauli, qd)
```

```python
import matplotlib.pyplot as plt
from Define_Paulis import Z,X,Mdot
import scipy.linalg as ln
from Define_Analytical import S_op, rho_op, FreeE

H = Z(0,2) + Z(1,2) + 0.3*Mdot([X(0,2),X(1,2)])

Ef = FreeE(T,H,rho_op(H,T))
Efl = [Ef for k in kl]

plt.plot(kl,El)
plt.plot(kl,Efl)
plt.ylim(-3.5,0.0)
```
