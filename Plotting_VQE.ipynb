```python
import pickle

alpha = 0.201
A = 10
a = 0.1
gamma = 0.101
c = 0.4

T=1
file_to_read = open("VQE_Gunlycke_T"+str(T)+".pickle", "rb")
out_1 = pickle.load(file_to_read)

T=2
file_to_read = open("VQE_Gunlycke_T"+str(T)+".pickle", "rb")
out_2 = pickle.load(file_to_read)

T=3
file_to_read = open("VQE_Gunlycke_T"+str(T)+".pickle", "rb")
out_3 = pickle.load(file_to_read)

T=4
file_to_read = open("VQE_Gunlycke_T"+str(T)+"_A_"+str(A)+str(T)+"_a_"+str(a)+str(T)+"_c_"+str(c)+str(T)+"_alpha_"+str(alpha)+str(T)+"_gamma_"+str(gamma)+".pickle", "rb")
out_4 = pickle.load(file_to_read)
```

```python
import matplotlib.pyplot as plt

El_1 = [out_1[i]['E'] for i in range(len(out_1))]
El_2 = [out_2[i]['E'] for i in range(len(out_2))]
El_3 = [out_3[i]['E'] for i in range(len(out_3))]
El_4 = [out_4[i]['E'] for i in range(len(out_4))]
E0 = [-3.34084762 for i in range(len(out_4))]
tl = [i for i in range(len(El_1))]
      
plt.title("VQE runs with differnt lengths of Anzatz")
plt.plot(tl,El_1,label = 'T = 1')
plt.plot(tl,El_2,label = 'T = 2')
plt.plot(tl,El_3,label = 'T = 3')
plt.plot(tl,El_4,label = 'T = 4')
plt.plot(tl,E0,label = 'Ground State')
plt.xlabel('Run Time')
plt.ylabel('Energy')
plt.legend()
plt.show()
```
